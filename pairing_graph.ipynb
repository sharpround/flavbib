```python
import networkx as nx
import pickle
import matplotlib.pyplot as plt

pairing_dict = pickle.load(open("pairing.pickle", "rb"))
```

```python
G = nx.Graph()
G.add_nodes_from(pairing_dict.keys())
```

```python
import re

out = open('pairings.txt', 'w')
for flavor in pairing_dict:
    for l in pairing_dict[flavor]:
        n1 = flavor
        n2 = re.split(':|, esp.|\u2014| \(|\(| and| /', l.lower())[0]
        G.add_edge(n1, n2)
```

```python
cliques = list(nx.enumerate_all_cliques(G))
```
